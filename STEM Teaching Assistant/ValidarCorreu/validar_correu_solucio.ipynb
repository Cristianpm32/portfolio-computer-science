# Validar adreça de correu

Comprovar si una adreça de correu electrònic és vàlida és una habilitat útil. Assegura't que una adreça de correu electrònic segueixi aquestes regles:

- Té un símbol @
- La part abans de @ no comença ni acaba amb . o -
- La part després de @ acaba en .com, .net, o .org
- També considerem que darrere de l'arrova hi ha d'haver un caràcter alfabètic, per exemple $...@e.com$
- Considerem que la mínima adreça vàlida pot ser de la forma $b@e.com$, que són 7 caràcters. És a dir, un caràcter vàlid davant de l'arrova, un vàlid després i .com, .net o .org

Si ho fa, digues 'Correcte'. Si no, digues 'Incorrecte'. 
Exemples: 

- $"test@example.com" -> "Correcte"$
- $"test@example.c" -> "Incorrecte"$
- $"-test@example.com" -> "Incorrecte"$
- $"test@.com" -> "Incorrecte"$
- $"test@example.net" -> "Correcte"$
- $"test.@example.com" -> "Incorrecte"$
```

````python
def validar_adreca(adreca):
    """
    Funció que comprova si una adreça de correu electrònic és vàlida. Per ser-ho, l'adreça ha de tenir com a mínim 7 caràcters, 
    començar per un caràcter que no sigui . o - i contenir una @, seguida d'almenys 5 caràcters i acabar en .com, .net o .org.

    Paràmetres:
    ___________
    adreca : str
        L'adreça de correu electrònic a comprovar
    Retorna:
    ________
    str
        Missatge que indica si l'adreça és correcte o incorrecte
    """
    llargada = len(adreca)
    
    # Comprovem que l'adreça no comenci per . o - i que tingui com a mínim 7 caràcters
    if adreca[0] == "." or adreca[0] == "-" or llargada < 7:       # Complexitat O(1)
        return "Incorrecte"
    else:
        i = 1           # Variable que ens servirà per recórrer l'adreça
        arrova = False  # Variable que ens indicarà si hi ha una @ a l'adreça

    # Comprovem que hi hagi una @ a l'adreça
        while not arrova and i < llargada:                          # Complexitat O(n)
            if adreca[i] == "@":                                    # Complexitat O(1)
                arrova = True
            i+=1                                                    # Complexitat O(1)

    # Si hi ha @, comprovem que darrere de l'arrova hi hagi com a mínim 5 caràcters i que acabi en .com, .net o .org
        if(arrova):
            if adreca[i-2] == "." or adreca[i-2] == "-": # Comprovem que darrere de l'arrova no hi hagi un . o un - . Complexitat O(1)
                return "Incorrecte"
            elif len(adreca[i:llargada]) < 5:            # Comprovem que hi hagi com a mínim 5 caràcters darrere de l'arrova. Complexitat O(1)
                return "Incorrecte"
            else:
                # Comprovem que l'adreça acabi en .com, .net o .org . Complexitat O(1)
                if adreca[llargada-4:llargada] != ".com" and adreca[llargada-4:llargada] != ".net" and adreca[llargada-4:llargada] != ".org":
                    return "Incorrecte"
        else:
            return "Incorrecte"
            
    # Finalment, si l'adreça és vàlida, retornem un missatge d'èxit
    return "Correcte"
````

# Complexitat del codi

La complexitat del codi és $O(n)$, on $n$ és la longitud de la cadena d'adreça. El càlcul de la longitud i les condicions inicials tenen complexitat $O(1)$, però el bucle `while` que busca la "@" pot recórrer tota la cadena, fent que el temps d'execució depengui de la mida de l'entrada. Les comprovacions després de trobar la "@" també són constants $O(1)$. Així, el cost principal prové del bucle, donant al codi una complexitat final de $O(n)$.


````python
# PROVA EL TEU CODI
assert validar_adreca("testexample.com") == "Incorrecte"
assert validar_adreca("test@examplecom") == "Incorrecte"
assert validar_adreca("test@example.net") == "Correcte"
assert validar_adreca(".test@example.com") == "Incorrecte"
assert validar_adreca("-test@example.com") == "Incorrecte"
assert validar_adreca("test@example.c") == "Incorrecte"
assert validar_adreca("test@example.net") == "Correcte"
assert validar_adreca("test@example.com") == "Correcte"
assert validar_adreca("test.@example.com") == "Incorrecte"
assert validar_adreca("te.st@exam-ple.org") == "Correcte"
assert validar_adreca("test@abc.de") == "Incorrecte"
````
