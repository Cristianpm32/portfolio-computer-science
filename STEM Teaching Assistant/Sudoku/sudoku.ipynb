# Sudoku Solver

El **Sudoku** és un trencaclosques de lògica on s'ha de completar una quadrícula de 9x9 amb nombres del 1 al 9, assegurant que:
1. Cada fila contingui els números del 1 al 9 sense repetir cap número.
2. Cada columna contingui els números del 1 al 9 sense repetir cap número.
3. Cada subquadrat de 3x3 (anomenat regió) contingui també els números del 1 al 9, sense repeticions.

Un Sudoku tradicional comença amb alguns números ja col·locats i l'objectiu és omplir la resta seguint les regles esmentades.

### Precondició:
- Tenim una quadrícula de 9x9 amb alguns números ja col·locats i espais buits que s'han d'omplir.

### Postcondició:
- Després de resoldre'l, cada fila, columna i subquadrat de 3x3 té exactament els números de l'1 al 9, sense repeticions.

```python
def imprimir_sudoku(matriu):
    """
    Aquesta funció imprimeix una matriu 9x9 com un Sudoku, amb separacions visuals clares entre els blocs 3x3.
    
    Paràmetres:
    -----------
    matriu : list
        Una llista de llistes de 9x9 que representa un Sudoku.

    Retorna:
    --------
    Imprimeix el Sudoku amb separacions visuals clares.
    """
    print("+-------+-------+-------+")  # Línia inicial per començar el Sudoku
    for i in range(9):
        if i % 3 == 0 and i != 0:
            print("+-------+-------+-------+")  # Línia més gruixuda per separar els blocs horitzontals 3x3

        for j in range(9):
            if j % 3 == 0:
                print("|", end=" ")  # Separació vertical entre blocs 3x3

            # Format amb espai per cada número, substituïm els 0 per espais buits
            if j == 8:
                print(f"{matriu[i][j] if matriu[i][j] != 0 else ' '} |")  # Final de la fila
            else:
                print(f"{matriu[i][j] if matriu[i][j] != 0 else ' '}", end=" ")
    
    # Línia final per tancar el Sudoku
    print("+-------+-------+-------+")
```
