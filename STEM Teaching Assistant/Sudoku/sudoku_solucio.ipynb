# Sudoku Solver

El **Sudoku** és un trencaclosques de lògica on s'ha de completar una quadrícula de 9x9 amb nombres del 1 al 9, assegurant que:
1. Cada fila contingui els números del 1 al 9 sense repetir cap número.
2. Cada columna contingui els números del 1 al 9 sense repetir cap número.
3. Cada subquadrat de 3x3 (anomenat regió) contingui també els números del 1 al 9, sense repeticions.

Un Sudoku tradicional comença amb alguns números ja col·locats i l'objectiu és omplir la resta seguint les regles esmentades.

### Precondició:
- Tenim una quadrícula de 9x9 amb alguns números ja col·locats i espais buits que s'han d'omplir.

### Postcondició:
- Després de resoldre'l, cada fila, columna i subquadrat de 3x3 té exactament els números de l'1 al 9, sense repeticions.

### Tasca a fer

La vostra feina consisteix en implementar un programa que permeti resoldre un Sudoku. Per això, com a paràmetre d'entrada es rebrà una matriu de 9x9 i la funció haurà de retornar una matriu amb el Sudoku resolt. Per poder realitzar això, es farà ús de l'algorisme de Backtracking



```python

import random

# Genera un tauler de sudoku amb un nombre aleatori de caselles buides
def genera_sudoku():
    base = 3
    costat = base * base

    # Patrons per a la base d'un Sudoku complet
    def patró(r, c): return (base*(r % base) + r//base + c) % costat
    def barrejar(s): return random.sample(s, len(s))

    # Barrejar files, columnes i números
    rBase = range(base)
    files = [g*base + r for g in barrejar(rBase) for r in barrejar(rBase)]
    columnes = [g*base + c for g in barrejar(rBase) for c in barrejar(rBase)]
    números = barrejar(range(1, base*base+1))

    # Crear el tauler complet
    tauler = [[números[patró(r, c)] for c in columnes] for r in files]

    # Eliminar alguns números per crear un tauler parcial
    quadrats = costat * costat
    buits = random.randint(40, 60)  # Controlar quants números eliminar
    for _ in random.sample(range(quadrats), buits):
        tauler[_//costat][_%costat] = 0

    return tauler
```

```python
def comprova_casella(tauler, casella, num):
    """
    Aquesta funció comprova si un número és vàlid per a una casella concreta

    Paràmetres:
    ------------
    tauler : llista de llistes
        Tauler de sudoku
    casella : tupla
        Coordenades de la casella a comprovar
    num : int
        Número a comprovar
    
    Retorna:
    ------------
    bool
        True si el número és vàlid, False si no ho és
    """
    costat = len(tauler)

    # Comprovar files
    for i in range(costat):
        if tauler[casella[0]][i] == num:
            return False

    # Comprovar columnes
    for i in range(costat):
        if tauler[i][casella[1]] == num:
            return False
    # Comprovar quadrat
    quadrat_fila = casella[0] // 3
    quadrat_columna = casella[1] // 3
    for i in range(quadrat_fila * 3, quadrat_fila * 3 + 3):
        for j in range(quadrat_columna * 3, quadrat_columna * 3 + 3):
            if tauler[i][j] == num:
                return False
    return True

def solve(tauler):
    """
    Aquesta funció resol un tauler de sudoku

    Paràmetres:
    ------------
    tauler : llista de llistes
        Tauler de sudoku
    
    Retorna:
    ------------
    bool
        True si s'ha resolt el tauler, False si no s'ha pogut
    """
    costat = len(tauler)

    # Buscar casella buida
    for i in range(costat):
        for j in range(costat):
            if tauler[i][j] == 0:

                # Provar números
                for num in range(1, 10):
                    
                    # Fem ús de la funció auxiliar per comprovar si el número és vàlid
                    if comprova_casella(tauler, (i, j), num):
                        tauler[i][j] = num
                        if solve(tauler):
                            return True
                        tauler[i][j] = 0
                return False
    return True
```

```python
def imprimir_sudoku(matriu):
    """
    Aquesta funció imprimeix una matriu 9x9 com un Sudoku, amb separacions visuals clares entre els blocs 3x3.
    
    Paràmetres:
    -----------
    matriu : list
        Una llista de llistes de 9x9 que representa un Sudoku.

    Retorna:
    --------
    Imprimeix el Sudoku amb separacions visuals clares.
    """
    print("+-------+-------+-------+")  # Línia inicial per començar el Sudoku
    for i in range(9):
        if i % 3 == 0 and i != 0:
            print("+-------+-------+-------+")  # Línia més gruixuda per separar els blocs horitzontals 3x3

        for j in range(9):
            if j % 3 == 0:
                print("|", end=" ")  # Separació vertical entre blocs 3x3

            # Format amb espai per cada número, substituïm els 0 per espais buits
            if j == 8:
                print(f"{matriu[i][j] if matriu[i][j] != 0 else ' '} |")  # Final de la fila
            else:
                print(f"{matriu[i][j] if matriu[i][j] != 0 else ' '}", end=" ")
    
    # Línia final per tancar el Sudoku
    print("+-------+-------+-------+")
```

```python
"""
Aquest codi genera un Sudoku aleatori, el resol i imprimeix el Sudoku original i el resolt.

"""
# Generem un Sudoku aleatori
sudoku = genera_sudoku()
# Cridem la funció per imprimir el Sudoku
print("Sudoku original:")
imprimir_sudoku(sudoku)
# Cridem la funció per resoldre el Sudoku
solve(sudoku)
# Cridem la funció per imprimir el Sudoku resolt
print("\nSudoku resolt:")
imprimir_sudoku(sudoku)
```
