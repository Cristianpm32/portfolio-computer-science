# FizzBuzz

FizzBuzz és un problema clàssic de programació on has d'imprimir números de l'1 al n, amb una sèrie de condicions.

Condicions:
- Si el nombre és divisible per 3, en comptes del número imprimeix "Fizz".
- Si el nombre és divisible per 5, en comptes del número imprimeix "Buzz".
- Si el nombre és divisible tant per 3 com per 5, en comptes del número imprimeix "FizzBuzz".

En tots els altres casos imprimeix el número directament.

Es demana implementar un programa que faci això i calcular la complexitat de les línies de codi i del programa en general.


```python
def fizzbuzz(n):
    """
    Imprimeix els números de l'1 al n amb les condicions de FizzBuzz.
    Parameters
    ----------
    n : int
        Número màxim a imprimir.
    Returns
    -------
    llista: list
        Llista amb els números de l'1 al n amb les condicions de FizzBuzz.
    """
    
    llista = [] # Complexitat: O(1)

    # Iterem sobre tots els números de l'1 al n
    for num in range(1, n + 1): # Complexitat: O(n)

    # Comprovem si el número és divisible per 3, 5
        if num % 15 == 0:  # Complexitat: O(1)
            llista.append("FizzBuzz")   # Complexitat: O(1)

    # Comprovem si el número és divisible per 3
        elif num % 3 == 0: # Complexitat: O(1)
            llista.append("Fizz")       # Complexitat: O(1)
    
    # Comprovem si el número és divisible per 5
        elif num % 5 == 0:              # Complexitat: O(1)
            llista.append("Buzz")       # Complexitat: O(1)
    
    # Si no és divisible per 3 ni per 5, imprimim el número
        else:              
            llista.append(num)          # Complexitat: O(1)
    return llista 
```

```python
assert fizzbuzz(0) == []
assert fizzbuzz(15) == [1, 2, "Fizz", 4, "Buzz", "Fizz", 7, 8, "Fizz", "Buzz", 11, "Fizz", 13, 14, "FizzBuzz"]
assert fizzbuzz(30) == [1, 2, "Fizz", 4, "Buzz", "Fizz", 7, 8, "Fizz", "Buzz", 11, "Fizz", 13, 14, "FizzBuzz", 16, 17, "Fizz", 19, "Buzz", "Fizz", 22, 23, "Fizz", "Buzz", 26, "Fizz", 28, 29, "FizzBuzz"]
assert fizzbuzz(2) == [1, 2]
assert fizzbuzz(10) == [1, 2, "Fizz", 4, "Buzz", "Fizz", 7, 8, "Fizz", "Buzz"]
```

# Complexitat

Observem que al bucle hi ha un bloc de condicionals, on cada possible acció aporta una complexitat de $O(1)$. Tenim $3n$ passos computacionals, per tant la complexitat del programa és de $O(n)$.
