# Shortest Remaining Time First (SRTF) Scheduling

## Descripció

El **Shortest Remaining Time First (SRTF)** és un algorisme de planificació de tasques preemptiu utilitzat en sistemes operatius per a la gestió de processos. En aquest mètode, el sistema sempre selecciona el procés que té el **temps restant d'execució més curt**. Si arriba un nou procés amb un temps estimat més curt que el procés actual en execució, aquest substitueix l'anterior.

Aquest algorisme es diferencia del **Shortest Job Next (SJN)** perquè permet la preempció, és a dir, un procés pot ser interromput en qualsevol moment si arriba un altre procés amb un temps de finalització més curt.

## Característiques

- **Preemptiu:** el procés en execució pot ser interromput si arriba un procés amb menys temps restant.
- **Eficiència:** tendeix a reduir el **temps d'espera mitjà** i el **temps de resposta** dels processos.
- **Desavantatges:** pot causar el problema de la **faminació** (starvation), on els processos llargs poden quedar bloquejats si continuen arribant processos més curts.
- **Complexitat:** cal recalcular contínuament el temps restant dels processos, la qual cosa fa que l'algorisme sigui més complex de gestionar que altres com el First Come First Serve (FCFS).

## Exemple

Si tenim els següents processos amb els seus temps d'arribada i durades:

| Procés | Temps d'arribada | Durada |
|--------|------------------|--------|
| P1     | 0                | 7      |
| P2     | 2                | 4      |
| P3     | 4                | 1      |
| P4     | 5                | 4      |

### Planificació segons SRTF

1. **Temps 0**: Només ha arribat **P1**, així que comença a executar-se.
2. **Temps 2**: Arriba **P2**, que té menys temps restant (4) que **P1** (li queden 5). **P1** és interromput i s'inicia **P2**.
3. **Temps 4**: Arriba **P3**, amb només 1 unitat de temps, per tant, s'interromp **P2** (li queden 2) i s'executa **P3**.
4. **Temps 5**: Finalitza **P3** i arriba **P4**. Com el temps d'execució restant de **P2** és inferior al de **P4**, es reprèn **P2** i es deixa **P4** a la cua d'espera
5. **Temps 7**: Finalitza **P2** i es reprèn **P4** (li queden 4),ja que el seu temps d'execució restant és inferior al de **P5** (li queden 5).
6. **Temps 11**: Finalitza **P4** i es reprèn **P1**
7. **Temps 16**: Finalitza **P1** i, per tant, ja han finalitzat tots els processos.

| Procés | Temps d'arribada | Durada | Temps de finalització | Temps de resposta | Temps d'espera |
|--------|------------------|--------|-----------------------|-------------------|----------------|
| P1     | 0                | 7      | 16                    | 16                 | 9              |
| P2     | 2                | 4      | 7                     | 5                 | 1              |
| P3     | 4                | 1      | 5                     | 1                 | 0              |
| P4     | 5                | 4      | 11                    | 6                 | 2             |

- **Temps de resposta**: és el temps que triga cada procés des del seu moment d'arribada fins que finalitza

- **Temps d'espera**: és el temps total que un procés està esperant en la cua, excloent el temps que està en execució.

- **Temps de finalització**: és el temps en què cada procés finalitza la seva execució.

### Càlcul del temps mitjà

- **Temps d'espera mitjà**: $\frac{9 + 1 + 0 + 2}{4} = 3.0$

- **Temps de resposta mitjà**: $\frac{16 + 5 + 1 + 6}{4} = 7.0$

- **Temps de finalització mitjà**: $\frac{16 + 7 + 5 + 11}{4} = 9.75$

Aquestes mètriques mostren com es comporta l'algorisme i ajuden a entendre com es distribueix l'execució de les tasques.
