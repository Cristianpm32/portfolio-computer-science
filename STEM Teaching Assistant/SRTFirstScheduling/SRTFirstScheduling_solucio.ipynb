# Shortest Remaining Time First (SRTF) Scheduling

## Descripció

El **Shortest Remaining Time First (SRTF)** és un algorisme de planificació de tasques preemptiu utilitzat en sistemes operatius per a la gestió de processos. En aquest mètode, el sistema sempre selecciona el procés que té el **temps restant d'execució més curt**. Si arriba un nou procés amb un temps estimat més curt que el procés actual en execució, aquest substitueix l'anterior.

Aquest algorisme es diferencia del **Shortest Job Next (SJN)** perquè permet la preempció, és a dir, un procés pot ser interromput en qualsevol moment si arriba un altre procés amb un temps de finalització més curt.

## Característiques

- **Preemptiu:** el procés en execució pot ser interromput si arriba un procés amb menys temps restant.
- **Eficiència:** tendeix a reduir el **temps d'espera mitjà** i el **temps de resposta** dels processos.
- **Desavantatges:** pot causar el problema de la **faminació** (starvation), on els processos llargs poden quedar bloquejats si continuen arribant processos més curts.
- **Complexitat:** cal recalcular contínuament el temps restant dels processos, la qual cosa fa que l'algorisme sigui més complex de gestionar que altres com el First Come First Serve (FCFS).

## Exemple

Si tenim els següents processos amb els seus temps d'arribada i durades:

| Procés | Temps d'arribada | Durada |
|--------|------------------|--------|
| P1     | 0                | 7      |
| P2     | 2                | 4      |
| P3     | 4                | 1      |
| P4     | 5                | 4      |

### Planificació segons SRTF

1. **Temps 0**: Només ha arribat **P1**, així que comença a executar-se.
2. **Temps 2**: Arriba **P2**, que té menys temps restant (4) que **P1** (li queden 5). **P1** és interromput i s'inicia **P2**.
3. **Temps 4**: Arriba **P3**, amb només 1 unitat de temps, per tant, s'interromp **P2** (li queden 2) i s'executa **P3**.
4. **Temps 5**: Finalitza **P3** i arriba **P4**. Com el temps d'execució restant de **P2** és inferior al de **P4**, es reprèn **P2** i es deixa **P4** a la cua d'espera
5. **Temps 7**: Finalitza **P2** i es reprèn **P4** (li queden 4),ja que el seu temps d'execució restant és inferior al de **P5** (li queden 5).
6. **Temps 11**: Finalitza **P4** i es reprèn **P1**
7. **Temps 16**: Finalitza **P1** i, per tant, ja han finalitzat tots els processos.

### Temps emprat i altres mètriques

Per calcular les mètriques de rendiment, hem de considerar el **temps d'execució**, el **temps d'espera** i el **temps de resposta** per a cada procés.

| Procés | Temps d'arribada | Durada | Temps de finalització | Temps de resposta | Temps d'espera |
|--------|------------------|--------|-----------------------|-------------------|----------------|
| P1     | 0                | 7      | 16                    | 16                 | 9              |
| P2     | 2                | 4      | 7                     | 5                 | 1              |
| P3     | 4                | 1      | 5                     | 1                 | 0              |
| P4     | 5                | 4      | 11                    | 6                 | 2             |

- **Temps de resposta**: és el temps que triga cada procés des del seu moment d'arribada fins que finalitza

- **Temps d'espera**: és el temps total que un procés està esperant en la cua, excloent el temps que està en execució.

- **Temps de finalització**: és el temps en què cada procés finalitza la seva execució.

### Càlcul del temps mitjà

- **Temps d'espera mitjà**: $\frac{9 + 1 + 0 + 2}{4} = 3.0$

- **Temps de resposta mitjà**: $\frac{16 + 5 + 1 + 6}{4} = 7.0$

- **Temps de finalització mitjà**: $\frac{16 + 7 + 5 + 11}{4} = 9.75$

Aquestes mètriques mostren com es comporta l'algorisme i ajuden a entendre com es distribueix l'execució de les tasques.

### La vostra tasca

La feina consisteix en implementar l'algorisme SRTF a partir d'una llista de llistes, on cada element de la llista que conté: 

- **Primer element:** El número de procés

- **Segon element:** El temps de ràfega, és a dir, el temps que triga el procés en executar-se

- **Tercer element:** El temps d'arribada a la cua de processos

Cal implementar una funció que retorni el temps d'espera mitjà i el temps de finalització mitjà.

```python
def shortest_remaining_time_first(processos, n):
    """
    Funció que implementa l'algorisme de planificació de processos Shortest Remaining Time First (SRTF)

    Paràmetres:
    ------------
    processos : llista
        Llista de processos a planificar. Cada procés és una llista [id, rafega, arribada]
    n : int
        Número de processos a planificar
    
    Retorna:
    ------------
    temps_espera_mitja : float
        Temps d'espera mitjà dels processos
    temps_finalitzacio_mitja : float
        Temps de finalització mitjà dels processos
    temps_resposta_mitja : float
        Temps de resposta mitjà dels processos
    """
    processos = sorted(processos, key=lambda x: x[0])  # Ordenem els processos per id
    temps_espera = [0] * n               # Temps d'espera per a cada procés
    temps_rafega = [0] * n               # Temps d'rafega per a cada procés
    temps_resposta = [0] * n             # Temps de resposta per a cada procés
    temps_finalitzacio = [0] * n         # Temps de finalitzacio per a cada procés

    # Copiem les rafegues dels processos
    for i in range(n):
        temps_rafega[i] = processos[i][1]
    completat = 0                           # Numero de processos completats
    temps = 0                               # Temps actual

    # Mentre no s'hagin completat tots els processos
    while completat != n:
        index = -1                         # Procés amb menor rafega
        minim = 999999999                  # Valor màxim
        for i in range(n):

            # Si el procés està disponible en aquest temps i la seva rafega és menor que la mínima
            if processos[i][2] <= temps and (temps_rafega[i] < minim and temps_rafega[i] > 0):
                minim = temps_rafega[i]    # Actualitzem el valor mínim
                index = i                  # Actualitzem l'índex del procés
        if index == -1:                    # Si no hi ha processos disponibles  en aquest temps incrementem el temps
            temps += 1
        else:
            temps_rafega[index] -= 1                                                        # Decrementem la rafega del procés

            # Si la rafega és 0, el procés ha acabat
            if temps_rafega[index] == 0:              
                completat += 1                                                              # Incrementem el número de processos completats
                temps_finalitzacio[index] = temps + 1                                       # Actualitzem el temps de finalització
                temps_resposta[index] = temps_finalitzacio[index] - processos[index][2]     # Actualitzem el temps de resposta
                temps_espera[index] = temps_resposta[index] - processos[index][1]           # Actualitzem el temps d'espera
                temps_rafega[index] = 999999999                                             # Actualitzem la rafega a un valor molt gran per a que no sigui el mínim
            temps += 1
    temps_espera_mitja = sum(temps_espera)/n                                                # Calculem temps d'espera mitja
    temps_finalitzacio_mitja = sum(temps_finalitzacio)/n                                    # Calculem temps de resposta mitja
    temps_resposta_mitja = sum(temps_resposta)/n                                            # Calculem temps de finalitzacio mitja
    return temps_espera_mitja, temps_finalitzacio_mitja, temps_resposta_mitja


    
```

```python
shortest_remaining_time_first([[1, 6, 2], [2, 2, 5],[3, 8, 1], [4, 3, 0], [5, 4, 4]], 5)
shortest_remaining_time_first([[1, 7, 0], [2, 4, 2],[3, 1, 4], [4, 4, 5]], 4)
shortest_remaining_time_first([[1, 5, 0], [2, 2, 2], [3, 3, 3], [4, 6, 5]], 4) 
shortest_remaining_time_first([[1, 5, 0], [2, 3, 1], [3, 2, 2], [4, 4, 3]], 4)
```
