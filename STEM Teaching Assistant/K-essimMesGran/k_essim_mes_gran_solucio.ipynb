# Troba el k-èssim element més gran


Imagina que tens una llista de números tota desordenada, i necessites trobar el k-èssim número més gran. Aquesta tasca pot semblar complicada, però amb els algorismes adequats, es pot resoldre de manera eficient.

Per tant, donada una llista desordenada, cal trobar el k-èssim element més gran de la llista. Implem,enteu el vostre programa utilitzant un algorisme de dividir-i-vèncer.

__IMPORTANT:__ No es pot fer ús de la funció sort() de Python per ordenar la llista



```python
# Versió amb llista de comprensions
def k_essim_mes_gran(llista, k):
    """
    Funció que retorna el k-èssim element més gran d'una llista
    Paràmetres:
    ___________
    llista : list
        Llista de la que volem trobar l'element més gran
    k : int
        Posició de l'element més gran que volem trobar
    Retorna:
    ________
    int
    pivot : int
        Retorna l'element més gran de la llista a la posició k
    """
    llargada = len(llista)                          # Calculem la llargada de la llista. Complexitat O()
    pivot = llista[llargada // 2]                   # Seleccionem un element com a pivot. Complexitat O(1)
    menors = [x for x in llista if x < pivot]       # Creem la llista amb els elements menors que el pivot. Complexitat O(n)
    majors = [x for x in llista if x > pivot]       # Creem la llista amb els elements majors que el pivot. Complexitat O(n)
    if k <= len(majors):                            # Comprova si el k-èssim element més gran es troba a la llista de majors
        return k_essim_mes_gran(majors, k)          # Complexitat de la funció
    if k > len(llista) - len(menors):               # Comprova si el k-èssim element més gran es troba a la llista de menors
        return k_essim_mes_gran(menors, k - (len(llista) - len(menors))) # Complexitat de la funció
    else:
        return pivot                                # Si no es compleixen les condicions anteriors, el pivot és el k-èssim element més gran
    
```

# Complexitat de la funció
El programa està basat en l'algorisme kSelect, ideal pel càlcul de la mediana. Per tant, la complexitat d'aquest programa és la mateixa que la de l'algorisme, que en el cas promig es pot demostrar que és $O(n)$ i en el pitjor dels casos $O(n^2)$

```python
# Versió sense llistes de comprensions
def k_essim_mes_gran(llista, k):
    """
    Funció que retorna el k-èssim element més gran d'una llista
    Paràmetres:
    ___________
    llista : list
        Llista de la que volem trobar l'element més gran
    k : int
        Posició de l'element més gran que volem trobar
    Retorna:
    ________
    int
    pivot : int
        Retorna l'element més gran de la llista a la posició k
    """
    llargada = len(llista)          # Calculem la llargada de la llista. Complexitat O(1)
    pivot = llista[llargada // 2]   # Calculem el pivot. Complexitat O(1)
    menors = []                     # Creem una llista buida per als menors al pivot. Complexitat O(1)
    majors = []                     # Creem una llista buida per als majors al pivot. Complexitat O(1)

    # Recorrem la llista i afegim els elements a la llista corresponent. Complexitat O(n)*O(1) = O(n)
    for x in llista:
        if x < pivot:                                                       # Complexitat O(1)
            menors.append(x)                                                # Complexitat O(1)
        elif x > pivot:                                                     # Complexitat O(1)
            majors.append(x)                                                # Complexitat O(1)

    # Comprovem si el k-èssim element més gran es troba a la llista de majors
    if k <= len(majors): 
        return k_essim_mes_gran(majors, k)                                  # Complexitat de la funció
    # Comprovem si el k-èssim element més gran es troba a la llista de menors
    if k > len(llista) - len(menors):
        return k_essim_mes_gran(menors, k - (len(llista) - len(menors)))    # Complexitat de la funció 
    else:
        return pivot # Si no es compleixen les condicions anteriors, el pivot és el k-èssim element més gran
```

# Complexitat de la funció
El programa està basat en l'algorisme kSelect, ideal pel càlcul de la mediana. Per tant, la complexitat d'aquest programa és la mateixa que la de l'algorisme, que en el cas promig es pot demostrar que és $O(n)$ i en el pitjor dels casos $O(n^2)$

```python
# PROVA EL TEU CODI  
llista = [100, 200, 50, 300, 150, 75, 300, 25, 400, 150, 175, 400]
assert k_essim_mes_gran(llista, 1) == 400  # El més gran
assert k_essim_mes_gran(llista, 2) == 400  # El segon més gran
assert k_essim_mes_gran(llista, 3) == 300  # El tercer més gran
assert k_essim_mes_gran(llista, 4) == 300  # El quart més gran
assert k_essim_mes_gran(llista, 5) == 200  # El cinquè més gran
assert k_essim_mes_gran(llista, 6) == 175  # El sisè més gran
assert k_essim_mes_gran(llista, 7) == 150  # El setè més gran
assert k_essim_mes_gran(llista, 8) == 150  # L'octau més gran
assert k_essim_mes_gran(llista, 9) == 100  # El novè més gran
assert k_essim_mes_gran(llista, 10) == 75  # El desè més gran
```
