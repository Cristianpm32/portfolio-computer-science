# Find the Longest Word

A partir d'una cadena de text o frase que es passa per paràmetre, cal retornar la llargada de la paraula més llarga. Per afegir dificultat aquest problema, tindrem les següents restriccions:
- Un número no forma part d'una cadena, i acaba amb aquesta -> ***Hol83quetal***
 
_Hol_ és una cadena, _83_ no és una cadena, _quetal_ és una cadena. La llargada de la paraula més llarga seria 6 (_quetal_).

- Qualsevol símbol o espai acaba amb una cadena. És a dir, qualsevol caràcter que no sigui una lletra -> ***Hola estudiants, això és Algorísmica!***

La funció hauria de retornar 11, ja que _estudiants_ no compta amb la coma i _Algorísmica_ (11 lletres) és la paraula més llarga i no compta amb el signe d'exclamació

```python
#SOLUCIÓ 
def longest_word(frase):
    """
    Retorna la llargada de la paraula més llarga de la frase
    Paràmetres:
    ___________
    frase : str
        La frase de la qual volem saber la llargada de la paraula més llarga
    Retorna:
    ________
    int
    llargada_maxima : int
        La llargada de la paraula més llarga de la frase
    """
    
    # Inicialitzem les variables que ens ajudaran a trobar la llargada de la paraula més llarga
    llargada_actual = 0 # 1
    llargada_maxima = 0 # 1

    # Recorrem la frase lletra a lletra
    for lletra in frase: # Complexitat O(n)
    
    # Si la lletra és una lletra incrementem la llargada de la paraula actual
        if lletra.isalpha() : # 1 + 1
            llargada_actual += 1 

    # Si lletra és un espai, comprovem si la llargada de la paraula actual és més gran que la llargada de la paraula més llarga fins ara
        else: # 1 + (1 + 1)
            if llargada_actual > llargada_maxima: 
                llargada_maxima = llargada_actual 

    # Reiniciem la llargada de la paraula actual
            llargada_actual = 0                  

    # En el cas que la paraula més llarga sigui la última de la frase o la frase només tingui una paraula
    if llargada_actual > llargada_maxima: # 1 + 1
        llargada_maxima = llargada_actual

    return llargada_maxima
```

# Complexitat
Es fan dues assignacions ($O(1)$) i un bucle on hi ha condicionals i una assignació de complexitat $O(1)$, per tant els passos computacionals del bucle són de $5n$. Seguidament, tenim un condicional fora del bucle que és de 2 passos computacionals Per tant, podem dir que els passos computacionals són de $5n +4$, és a dir una complexitat en notació O gran de $O(n)$

```python
# Cas de prova 1: Una sola paraula
assert longest_word("Hola") == 4

# Cas de prova 2: Diverses paraules amb diferents longituds
assert longest_word("El ràpid gos marró salta per sobre del gos mandrós") == 7

# Cas de prova 3: Diverses paraules amb la mateixa longitud
assert longest_word("El gat i el barret") == 6

# Cas de prova 4: Cadena buida
assert longest_word("") == 0

# Cas de prova 5: Paraules d'un sol caràcter
assert longest_word("A B C D E F G H I") == 1

# Cas de prova 6: Paraules amb espais al principi i al final
assert longest_word("  Espais  ") == 6

# Cas de prova 7: Paraules amb caràcters especials
assert longest_word("Hola,món!") == 4

# Cas de prova 8: Paraules amb números
assert longest_word("La resposta és 42") == 8

# Cas de prova 9: Paraules amb majúscules i minúscules
assert longest_word("HeLLo MoN") == 5

assert longest_word("Hola estudiants, això és Algorísmica!") == 11

```
