# Conway Sequence

La seqüència de Conway, o més coneguda com la seqüència "Look-and-Say", és una seqüència de nombres que s'obté descrivint oralment (o "llegint") la seqüència anterior. És un exemple molt interessant d'un patró numèric que evoluciona de forma aparentment simple, però que pot generar resultats sorprenents.

L'element clau d'aquesta seqüència és que cada nombre es llegeix en funció de la quantitat de dígits que conté el nombre anterior. Aquesta seqüència va ser popularitzada pel matemàtic John H. Conway, d'aquí el seu nom.

Com funciona la seqüència?
El primer nombre de la seqüència és 1. Des d'aquí, cada pas es construeix descrivint el nombre anterior.

Vegem un exemple detallat:

- 1 → el llegim com "un 1", per tant el següent nombre és 11.
- 11 → el llegim com "dos 1s", per tant el següent nombre és 21.
- 21 → el llegim com "un 2, un 1", per tant el següent nombre és 1211.
- 1211 → el llegim com "un 1, un 2, dos 1s", per tant el següent nombre és 111221.
- 111221 → el llegim com "tres 1s, dos 2s, un 1", per tant el següent nombre és 312211.

I així successivament. A partir d'aquí, la seqüència creix en complexitat, però el mètode per generar-la sempre és el mateix: descriure el nombre anterior.

Nosaltres farem una modificació d'aquesta seqüència, on es podrà modificar el primer nombre. Es demana implementar una funció que donats un primer nombre $r$ i un número de línia $n$, retorni la n-èssima línia.
Per exemple, si donem per paràmetres (1, 6), haurà de retornar 312211

```python
def conway_sequence(r,n):
    """
    Aquesta funció retorna la n-èssima seqüència de Conway a partir del primer terme r.

    Paràmetres:
    ____________
    r: int
        El primer terme de la seqüència.
    n: int
        La posició de la seqüència que volem calcular.
    
    Retorn:
    _____________
    str
        La n-èssima seqüència de Conway separada per espais.
    """


    # Afegim el primer element de la seqüència. El convertim a string perquè haurem de treballar amb strings.
    seq = [[str(r)]]

    # Anem afegint elements a la seqüència fins a arribar a n.
    for _ in range(1, n): # Fem servir _ perquè no ens importa el valor de la variable. Complexitat O(n)* O(m), on m és el nombre de nombres de la seqüència anterior.

        # Agafem l'últim element de la seqüència, que és una llista. Creem una llista buida per al nou element.
        # També creem un comptador per saber quants cops es repeteix un nombre.
        anterior = seq[-1]                                                      # Complexitat O(1)
        nou_terme = []                                                          # Complexitat O(1)
        comptador = 1                                                           # Complexitat O(1)

        # Recorrem la llista anterior per poder crear el nou element.
        for i in range(1, len(anterior)):                                       # Complexitat O(m)

            # Si el nombre actual és igual al nombre anterior, incrementem el comptador.
            if anterior[i] == anterior[i-1]:                                    # Complexitat O(1)
                comptador += 1

            # Si no, afegim el comptador i el nombre anterior a la llista del nou element. Reiniciem el comptador.
            else:
                nou_terme.append(str(comptador))                                # Complexitat O(1)
                nou_terme.append(str(anterior[i-1]))                            # Complexitat O(1)
                comptador = 1                                                   # Complexitat O(1)

        # Afegim l'últim comptador i l'últim nombre a la llista del nou element. Afegim el nou element a la seqüència.
        nou_terme.append(str(comptador))                                        # Complexitat O(1)
        nou_terme.append(str(anterior[-1]))                                     # Complexitat O(1)
        seq.append(nou_terme)                                                   # Complexitat O(1)
    
    # Retornem l'últim element de la seqüència, que és la n-èssima seqüència de Conway en forma de string separat per espais.
    return " ".join(seq[-1])

    
```

# Complexitat del codi

La complexitat del codi podria ser $O(n)$ si la longitud dels termes fossin prou menors a $n$. No obstant això, hi poden haver casos on la longitud d'aquests termes de la seqüència poden crèixer molt ràpid. Per tant, si diem $m$ a la longitud del terme més gran, tindríem una complexitat de $O(n) · O(m) -> O(n·m)$. Com ens hem de posar en el pitjor dels casos, la complexitat d'aquest codi és de $O(n·m)$
