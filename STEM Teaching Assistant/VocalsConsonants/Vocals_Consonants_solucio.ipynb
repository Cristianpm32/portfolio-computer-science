# Vocals i consonants
En aquest programa es demana implementar una funció que, donada una cadena de caràcters de només lletres minúscules i sense espais, retorni una llista amb la mateixa llargada de la cadena, però amb les següents característiques:
- Si una lletra de la cadena és una vocal, s'afegeix aquesta en majúscula.
- En cas que sigui una consonant, s'afegeix una Z majúscula.

Per exemple, si li passem la cadena "qoejnoiw", la funció hauria de retornar [Z, O, E, Z, Z, O, I, Z]


```python
    
def vocals_consonants(cadena):
           
    """
    Aquesta funció pren una cadena de text en minúscules com a entrada i retorna una llista on:
    - Les consonants de la cadena original es substituïxen per la lletra "Z" majúscula.

    Paràmetres:
    -------------------
    cadena (str): La cadena de text que es vol processar.

    Retorn:
    -------------------
    list: Una llista de caràcters on les vocals estan en majúscules i les consonants són substituïdes per "Z".
    """
    llista = [] # Creem una llista buida                           # Complexitat: O(1)

    # Recorrem la cadena lletra a lletra
    for lletra in cadena:                                          # Complexitat: O(n)
                          
        # Si la lletra és una vocal, l'afegim en majúscules        # Complexitat: O(1)
        if lletra in "aeiou":                                      # Complexitat: O(1)
            llista.append(lletra.upper())                          # Complexitat: O(1)
        # Si la lletra és una consonant, afegim una "Z" majúscula  # Complexitat: O(1)
        else:
            llista.append("Z")                                     # Complexitat: O(1)

    return llista                                                  
```

```python
# PROVA EL CODI
assert vocals_consonants("aeiou") == ["A", "E", "I", "O", "U"]
assert vocals_consonants("bcdfg") == ["Z", "Z", "Z", "Z", "Z"]
assert vocals_consonants("xyz") == ["Z", "Z", "Z"]
assert vocals_consonants("zxcvbnmqwe") == ["Z", "Z", "Z", "Z", "Z", "Z", "Z", "Z", "Z", "E"]
assert vocals_consonants("qwertyuiopasdfghjklzxcvbnm") == ["Z", "Z", "E", "Z", "Z", "Z", "U", "I", "O", "Z", "A", "Z", "Z", "Z", "Z", "Z", "Z", "Z", "Z", "Z", "Z", "Z", "Z", "Z", "Z", "Z"]
assert vocals_consonants("abcdefghijklmnopqrstuvwxyz") == ["A", "Z", "Z", "Z", "E", "Z", "Z", "Z", "I", "Z", "Z", "Z", "Z", "Z", "O", "Z", "Z", "Z", "Z", "Z", "U", "Z", "Z", "Z", "Z", "Z"]
```

# Complexitat del codi
Tenim per una banda el bucle for, que té una complexitat equivalent a la llargada de la cadena, per tant $O(n)$. Per altra banda, tenim que comprovar si és una vocal equival veure que a la cadena $"aeiou"$ no hi és la lletra. Com la llargada d'aquesta cadena no varia, la seva complexitat és constant, per tant $O(1)$. Per tant, la complexitat del programa és de $O(n)$
