# Suma dels múltiples de 3 o 5 menors a 1000

Els nombres naturals menors que 10 que són múltiples de 3 o 5 són 3, 5, 6 i 9. La seva suma és 23. Troba la suma de tots els nombres naturals menors que 1000 que són múltiples de 3 o 5.

```python
def n_iteracions_estricte(limitador, num):
   
    if (limitador % num == 0):
        iteracions = limitador//num -1
    else:
        iteracions = limitador //num
    return iteracions
```

```python
def multiples():
    """
    Aquesta funció retorna la suma de tots els naturals que 
    són múltiples de 3 o 5 i que son menors de 1000.
    
    Returns
    -------
    suma : int
    """
    iteracions3 = n_iteracions_estricte(1000,3)
    iteracions5 = n_iteracions_estricte(1000,5)
    iteracions15 = n_iteracions_estricte(1000,15)
    suma = 3*iteracions3*(iteracions3+1)//2 + 5*iteracions5*(iteracions5+1)//2 - 15*iteracions15*(iteracions15+1)//2
    return suma
```

```python
assert multiples() == 233168
```
